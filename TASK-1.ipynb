PROBABILITY USING BAYES THEOREM

```python
import pandas as pd



# Define conditions
passed = df['exam_score'] >= 50
study_gt_2 = df['study_hours_per_day'] > 2

# Compute probabilities
P_passed = passed.mean()
P_study_gt_2 = study_gt_2.mean()
P_passed_given_study_gt_2 = df[study_gt_2]['exam_score'].ge(50).mean()

# Apply Bayes' Theorem
P_study_gt_2_given_passed = (P_passed_given_study_gt_2 * P_study_gt_2) / P_passed

print(f"Probability that a student studied >2 hours given they passed: {P_study_gt_2_given_passed:.3f}")
```
